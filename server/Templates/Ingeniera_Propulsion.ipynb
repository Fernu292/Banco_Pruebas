# Reporte de banco de pruebas 

Reporte de analisis exploratorio de datos de la curva de empuje para el cohete speedey 1

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('../Data/PruebaBanco_Speedy2.cvs')
```

```python
data = df['Empuje']
data = np.array(data)
```

```python
plt.plot(data)
plt.xlabel("Time [dm]")
plt.ylabel("Empuje [N]")
```

```python
data.min()
```

### Filtrando datos de interferecia

Comenzamos a filtrar los datos de interferencia generados por pulsos electricos en el banco de pruebas

```python
dataAux = np.array([x for x in data if x>=0])
```

```python
plt.plot(dataAux)
plt.xlabel("Tiempo [ds]")
plt.ylabel("Empuje [N]")
plt.legend(["Empuje"])
```

```python
print(f'Empuje maximo: {dataAux.max()} N')
```

```python
string_cabecera = "MS1-41 15 120 P 0.02 0.060 Antares"
data_dict = {string_cabecera: dataAux}
df = pd.DataFrame(data_dict)
```
